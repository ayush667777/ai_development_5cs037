```
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
add_xy = x + y
print("3.1 Add x and y:\n", add_xy)

# 2. Subtract the two arrays
sub_xy = x - y
print("\n3.2 Subtract y from x:\n", sub_xy)

# 3. Multiply the array with integers of choice (say multiply by 3)
mul_x = x * 3
print("\n3.3 Multiply x by 3:\n", mul_x)

# 4. Find the square of each element in x
square_x = x ** 2
print("\n3.4 Square of each element in x:\n", square_x)

# 5. Dot product between v and w; x and v; x and y
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print("\n3.5 Dot product v·w:", dot_vw)
print("Dot product x·v:\n", dot_xv)
print("Dot product x·y:\n", dot_xy)

# 6. Concatenate x and y along row (axis=0) and concatenate v and w along column (axis=0 is only for 2D, so use np.vstack for 1D vectors)
concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.vstack((v, w))
print("\n3.6 Concatenate x and y along rows:\n", concat_xy_row)
print("Concatenate v and w along columns:\n", concat_vw_col)

# 7. Concatenate x and v - this will error due to shape mismatch. Catch and explain.
try:
    concat_xv = np.concatenate((x, v))
except ValueError as e:
    error_message = str(e)
print("\n3.7 Concatenate x and v results in error:")
print(error_message)
print("Explanation: Cannot concatenate arrays with different dimensions/shapes.")
```

```
# Problem 4 — Matrix Operations

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove that A * A^-1 = I
A_inv = np.linalg.inv(A)
result1 = A @ A_inv
print("4.1 A · A^-1:\n", result1)

# 2. Show that AB != BA
AB = A @ B
BA = B @ A
print("\n4.2 AB:\n", AB)
print("BA:\n", BA)
print("AB == BA?", np.array_equal(AB, BA))

# 3. Prove that (AB)^T = B^T A^T
lhs = (A @ B).T
rhs = B.T @ A.T
print("\n4.3 (AB)^T:\n", lhs)
print("B^T A^T:\n", rhs)
print("Check equality:", np.array_equal(lhs, rhs))
```
